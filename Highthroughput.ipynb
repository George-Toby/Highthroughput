Load required libraries and set up the environment

```python
import os
import subprocess

# Set up the environment
GENOME_INDEX = '/path/to/genome/index'  # Replace with your genome index
GENOME_SIZE = 'mm10'  # Replace with your genome size (e.g., mm10, hg38)
FASTQ_FILE = 'input.fastq.gz'  # Replace with your input FASTQ file
```

Quality control and trimming

```python
# Run FastQC
subprocess.run(['fastqc', FASTQ_FILE])

# Trim adapters and low-quality bases using Trimmomatic
subprocess.run(['trimmomatic', 'PE', '-phred33', FASTQ_FILE, 'output_trimmed.fastq.gz', 'ILLUMINACLIP:TruSeq3-PE.fa:2:30:10', 'LEADING:3', 'TRAILING:3', 'SLIDINGWINDOW:4:15', 'MINLEN:36'])
```

Read alignment using Bowtie

```python
!apt-get update && apt-get install -y bowtie2
# Align trimmed reads to the reference genome using Bowtie
subprocess.run(['bowtie2', '-x', GENOME_INDEX, '-p', '2', '-U', 'output_trimmed.fastq.gz', '-S', 'output.sam'])
```

SAM to BAM conversion and sorting

```python
!apt-get update && apt-get install -y samtools
!apt-get update && apt-get install -y bedtools
# Convert SAM to BAM using SAMtools
subprocess.run(['samtools', 'view', '-S', '-b', 'output.sam', '>', 'output.bam'])

# Sort the BAM file by coordinate using SAMtools
subprocess.run(['samtools', 'ort', 'output.bam', '-o', 'output.sorted.bam'])
```

Peak calling using MACS2

```python
!pip install --no-deps MACS2
import subprocess
GENOME_SIZE = 'hs'  # Replace with your actual genome size
subprocess.run(['macs2', 'callpeak', '-t', 'output.sorted.bam', '-f', 'BAM', '-g', GENOME_SIZE, '-n', 'output_peaks'])
```

Peak annotation using BEDTools

```python
# Annotate peaks with gene information using BEDTools
subprocess.run(['bedtools', 'intersect', '-a', 'output_peaks.bed', '-b', 'genes.bed', '-u', '>', 'output_peaks.annotated.bed'])
```
